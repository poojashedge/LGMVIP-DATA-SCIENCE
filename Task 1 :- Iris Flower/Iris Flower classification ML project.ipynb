## 1. Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("iris.csv")
data.head()
```

```python
data_headers=['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']
data.to_csv("iris.csv",header=data_headers,index=False)
ndata=pd.read_csv("iris.csv")
ndata.head()
```

## 4.Plotting

```python
sns.pairplot(data=ndata,hue='Species')
plt.show()
```

## 5.Comparision Of Species With The Feature

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.barplot(x='Species',y='SepalLength',data=ndata)
plt.subplot(2,2,2)
sns.barplot(x='Species',y='SepalWidth',data=ndata)
plt.subplot(2,2,3)
sns.barplot(x='Species',y='PetalLength',data=ndata)
plt.subplot(2,2,4)
sns.barplot(x='Species',y='PetalWidth',data=ndata)
```

## 6.Defining The Dependent And Indeopendent Variable

```python
X=ndata.drop(columns="Species")
y=ndata["Species"]
```

## 7. Split Into Training And Testing Datasets

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

## 8. Building Model

```python
from sklearn.ensemble import RandomForestClassifier 
classifier=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
classifier.fit(X_train,y_train)
```

## 9.Making The Confusion Matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score
y_pred=classifier.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
